# Multiple Linear Regression

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset = pd.read_csv("50_Startups.csv")
dataset
X = dataset.iloc[ : , : -1]
y = dataset.iloc[ : , -1]
```

```
print(dataset)
```

```
print(X)
```

```
print(y)
```

## Encoding categorical data

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[3])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Training the Multiple Linear Regression model on the Training set

* the multiple linear regression class will automatically take care of the **dummy variable traps** , so you don't have to delete any columns yourself .

* the multiple linear regression class will automatically take care of **variables that have the highest p value and make us the most accurate model** .





```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

Since there are multiple features, we will not be able to visualise our results in a graph. 
so we will directly look at the results.

```
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)

#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

to concatenate both y_pred and y_test, they need to be of the same shape.


```
y_pred_1 = y_pred.reshape(len(y_pred),1)
y_pred_1
```

```
y_test_1 = y_test.values.reshape(len(y_test),1)
y_test_1
```
